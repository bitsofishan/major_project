# Adding Data in CSV


```python
import numpy as np
import cv2
import os
from csv import writer
flag=1
def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)

directory=r"C:\Users\Ishan\Documents\ognc\train" #Directory in which files are stores in your computer
image_files=[os.path.join(directory,file) for file in  os.listdir(directory) if file.endswith(".jpg")]
#image_files.sort()
cx_square=cy_square=cx_hexa=cy_hexa=cx_sept=cy_sept=cx_pent=cy_pent=cx_triangle=cy_triangle=0
for i in image_files:
    img = cv2.imread(i)
    img = cv2.resize(img, (0, 0), fx = 0.8, fy =0.5) 
    imgGrey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   
    
    _, thrash = cv2.threshold(imgGrey, 127, 255, cv2.THRESH_BINARY)
    #thrash=cv2.Canny(thrash,255,255)
    
    
    #print(img)

    contours, _ = cv2.findContours(thrash, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)


    for contour in contours:
        approx = cv2.approxPolyDP(contour, 0.02* cv2.arcLength(contour, True), True)
        cv2.drawContours(img, [approx], 0, (0, 0, 0), 5)
        x = approx.ravel()[0]
        y = approx.ravel()[1] - 5
        area = cv2.contourArea(contour)
    
        if area>2100 and area<10000:
            if len(approx) == 3:
           
             M = cv2.moments(contour)
             cx_triangle = int(M['m10']/M['m00'])
             cy_triangle = int(M['m01']/M['m00'])
             pos=str(cx_triangle)+","+str(cy_triangle)
             cv2.putText(img,pos,(cx_triangle,cy_triangle),cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
            
            elif len(approx) == 4:
           
             M = cv2.moments(contour)
             cx_square = int(M['m10']/M['m00'])
             cy_square = int(M['m01']/M['m00'])
             pos=str(cx_square)+","+str(cy_square)
             cv2.putText(img,pos,(cx_square,cy_square),cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
            
            elif len(approx) == 5:
             M = cv2.moments(contour)
             cx_pent = int(M['m10']/M['m00'])
             cy_pent = int(M['m01']/M['m00'])
             pos=str(cx_pent)+","+str(cy_pent)
             cv2.putText(img,pos,(cx_pent,cy_pent),cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
            elif len(approx) == 6:
             M = cv2.moments(contour)
             cx_hexa = int(M['m10']/M['m00'])
             cy_hexa = int(M['m01']/M['m00'])
             pos=str(cx_hexa)+","+str(cy_hexa)
             cv2.putText(img,pos,(cx_hexa,cy_hexa),cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
            elif len(approx) == 7:
             M = cv2.moments(contour)
             cx_sept = int(M['m10']/M['m00'])
             cy_sept = int(M['m01']/M['m00'])
             pos=str(cx_sept)+","+str(cy_sept)
             cv2.putText(img,pos,(cx_sept,cy_sept),cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
            else:
             cv2.putText(img, "", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
    
    flag=str(flag)
    cv2.putText(img, flag, (800, 50), cv2.FONT_HERSHEY_COMPLEX, .5, (0, 0, 0))
    img=cv2.line(img,(cx_square,cy_square),(cx_triangle,cy_triangle),(255,0,0),1)
    img=cv2.line(img,(cx_square,cy_square),(cx_hexa,cy_hexa),(255,0,0),1)
    img=cv2.line(img,(cx_square,cy_square),(cx_sept,cy_sept),(255,0,0),1)
    img=cv2.line(img,(cx_square,cy_square),(cx_pent,cy_pent),(255,0,0),1)
    #cv2.imshow("shapes", img)
    cv2.imshow("shapes", thrash)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    new_row=[cx_square,cy_square,cx_hexa,cy_hexa,cx_sept,cy_sept,cx_pent,cy_pent,cx_triangle,cy_triangle]
    print(new_row)
    flag=int(flag)
    flag=flag+1
    # Append a list as new line to an old csv file
    append_list_as_row('train_data.csv', new_row)
    cx_square=cy_square=cx_hexa=cy_hexa=cx_sept=cy_sept=cx_pent=cy_pent=cx_triangle=cy_triangle=0
```
